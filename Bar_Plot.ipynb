```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data={
  "Year":[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020],
  "Sales":[100,150,200,250,300,350,400,450,500,550,600]
}
df = pd.DataFrame(data)
df
```

```python
#using MatPlotlib
plt.figure(figsize=(10, 5)) 
plt.plot(df["Year"],df["Sales"],color="red")
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
#using seaborn
plt.figure(figsize=(10,5))
sns.barplot(x="Year", y="Sales", data=df)
```

```python
#line plot using matplotlib
plt.plot(df["Year"],df["Sales"],color="red", linestyle="-", marker="^")
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
sns.lineplot(x="Year", y="Sales", data=data, linestyle="--", marker="o")
plt.title("Sales By Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
#BoxPlot
math_scores=[65,70,75,80,85,90,95,100]
english_scores=[55,60,65,70,75,80,85,90]

plt.figure(figsize=(10,5))
plt.boxplot([math_scores,english_scores],labels =['Math','English'])
plt.title('Boxplot of Math and English Scores')
plt.xlabel('Subject')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
#uisng seaborn
df=pd.DataFrame({
  'subject': ["Math"]*8+["Science"]*7,
  'score':[88,92,80,89,100,76,95,85,93,30,40,60,70,80,55]
})
sns.boxplot(x="subject", y="score", data=df)
plt.title('Seaborn Box plot')
plt.show()
```

```python
import numpy as np
data = np.array([[30, 20, 40],
                 [25, 35, 45],
                 [20, 25, 45]])

df = pd.DataFrame(data, columns=["Math", "Science", "English"],
                  index=["class_a", "class_b", "class_c"])

# Using matplotlib to plot heatmap
plt.imshow(data, cmap='YlGnBu')
plt.title('Heatmap')
plt.xticks([0, 1, 2], df.columns)
plt.yticks([0, 1, 2], df.index.tolist())
plt.colorbar(label='Scores')
plt.show()
```

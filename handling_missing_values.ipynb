```python
import numpy as np
import pandas as pd

data = {
    "name": ["Ram", "Shyam", "Hari", "Sita", "Gita"],
    "age": [20, np.nan, 22, np.nan, 24],
    "gender": ["M", "M", "M", "F", "F"],
    "city": ["Kathmandu", "Lalitpur", "Bhaktpur", "lalitpur", "Kathmandu"],
    "marks": [85, np.nan, 95, np.nan, 85]
}
df_names = pd.DataFrame(data)
```

```python
# Null values, null values count, drop null rows, fill null value with 20, drop duplicates

import pandas as pd

# Example DataFrame with null values
data = {
    'A': [1, 2, None, 4],
    'B': [None, 2, 3, 4],
    'C': [1, None, None, 4]
}
df = pd.DataFrame(data)
```

```python
# Check for null values (True/False mask)
print(df.isnull())
```

```python
# Count total null values in each column

print(df.isnull().sum())
```

```python
# Drop rows with any null values

df_dropped = df.dropna()
print(df_dropped)
```

```python
# Fill null values with 20

df_filled = df.fillna(20)
print(df_filled)
```

```python
# drop duplicates using for 
df_names.drop_duplicates(subset=["city", "marks"])
```

```python
# create 1d, 2d, 3d numpy array using the list

import numpy as np

# 1D list
list_1d = [1, 2, 3, 4, 5]
array_1d = np.array(list_1d)

print("1D Array:")
print(array_1d)
```

```python
# 2D list (list of lists)
list_2d = [[1, 2, 3], [4, 5, 6]]
array_2d = np.array(list_2d)

print("\n2D Array:")
print(array_2d)
```

```python
# 3D list (list of lists of lists)
list_3d = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
array_3d = np.array(list_3d)

print("\n3D Array:")
print(array_3d)
```

```python
# also apply operations such as mean, median, squareroot(sqrt)

# 1d mean, median squareroot(sqrt)

import numpy as np

list_1d = [1, 2, 3, 4, 5]
array_1d = np.array(list_1d)

print("1D Array:", array_1d)
print("Mean:", np.mean(array_1d))
print("Median:", np.median(array_1d))
print("Square Root:", np.sqrt(array_1d))
```

```python
# 2d 1d mean, median squareroot(sqrt)

list_2d = [[1, 2, 3], [4, 5, 6]]
array_2d = np.array(list_2d)

print("\n2D Array:\n", array_2d)
print("Mean:", np.mean(array_2d))
print("Median:", np.median(array_2d))
print("Square Root:\n", np.sqrt(array_2d))
```

```python
# 3d 1d mean, median squareroot(sqrt)

list_3d = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
array_3d = np.array(list_3d)

print("\n3D Array:\n", array_3d)
print("Mean:", np.mean(array_3d))
print("Median:", np.median(array_3d))
print("Square Root:\n", np.sqrt(array_3d))
```
